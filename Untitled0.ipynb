```python
# Load the csv file
import pandas as pd
df = pd.read_csv('films.csv')
```

```python
# Check the possible null value(s)
df.isnull()
```

```python
# Get the sum of null values for each column
df.isnull().sum()
```

```python
# Get data statistics such as count, mean, std, etc.
df.describe()
```

```python
# Determine the average popularity for each movie subject
subject_pop = df.groupby('Subject').Popularity.mean()
subject_pop
```

```python
# Identify the top 3 most popular subjects
top_subjects = subject_pop.sort_values(ascending=False).head(3)
top_subjects
```

```python
# Remove the records having nulls in Length, Subject, Populairty and Actor
df.dropna(subset=['Length', 'Subject', 'Popularity', 'Actor'])
```

```python
# Replace nulls in Directory columns with 'Unknown'
df['Actress'] = df['Actress'].replace('\\N', 'Unknown')
df
```

```python
# Replace nulls in Directory columns with 'Unknown'
df['Director'] = df['Director'].replace('\\N', 'Unknown')
df
```

```python
# Explore the trend of average movie lengths over the years between 1980 and 2000 with award staus breakdown
df.columns
```

```python
# Filter the movies based on the 'Year' column from 1980 to 2000
df_1980_2000 = df[(df['Year'] > 1979) & (df['Year'] < 1999)]
df_1980_2000
```

```python
# Group the data by 'Year' and 'Awards' to calculate the average 'Length'
df_1980_2000_awards = df_1980_2000.groupby(['Year', 'Awards']).Length.mean().reset_index()
df_1980_2000_awards
```

```python
import plotly.express as px
```

```python
fig = px.line(df_1980_2000_awards, x='Year', y = 'Length', color = 'Awards')
fig
```
